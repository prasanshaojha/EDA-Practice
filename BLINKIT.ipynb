```python
TOTAL_SALES  = df['Sales'].sum()

Average_sales = df['Sales'].mean()

total_item_sold = df['Sales'].count()

Avg_rating = df['Rating'].mean()

print(f'Total Sales : ${TOTAL_SALES:,.0f}')
print(f" AVERAGE SALES: ${Average_sales:,.0f}")
print(f"Total Item Sold: ${total_item_sold:,.0f}")
print(f"Averge Rating: $ {Avg_rating:,.0f}")
```

```python
Sales_per_Outlet = df.groupby('Outlet Type')['Sales'].count()

ax = Sales_per_Outlet.plot(kind = 'bar', color = 'blue' , edgecolor = 'black')
plt.title("Sales Count Per Outlet")
plt.xlabel("Outlet Type")
plt.ylabel("Number of Sales")
for i, v in enumerate(Sales_per_Outlet):
    ax.text(i, v+5, str(v), ha = "center", va = "bottom" )
plt.show()
```

```python
Sales_per_item_type = df.groupby('Item Type')['Sales'].sum()

ax = Sales_per_item_type.plot( kind = 'bar' , color = 'skyblue' , edgecolor = 'black')
plt.title("Total Sales Amount Per Item")
plt.xlabel("Item Type")
plt.ylabel("Sales Amount")

for i, v in enumerate(Sales_per_item_type):
    ax.text(i, v+ 6, str(v), ha = "center" , va= "bottom" )
plt.show()
```
